```python
path1 = '/home/liyouru/DataSet/JDD/Loan_Forecasting.csv'
path2 = '/home/liyouru/DataSet/JDD/submit_benchmark_.csv'

order = []
with open(path2, 'rb') as f:
    f.readline()
    for line in f:
        line = line.strip().split(',')
        order.append(str(line[0]))
        
m = {}
with open(path1, 'rb') as f:
    f.readline()
    for line in f:
        line = line.strip().split(',')
        m[line[0]] = max(0,float(line[1]))
        
with open('/home/liyouru/test_submit.csv', 'wb') as f:
    for x in order:
        f.write('{},{}\n'.format(x, m[x]))
```

```python
import pandas as pd
import numpy as np
path_loan = '/home/liyouru/DataSet/JDD/t_loan_sum.csv'
df = pd.read_csv(path_loan , header = 0, sep = ',')
print df.shape
print df.head()

value_uid = []
uid2value = {}
for each in df.iterrows():
    value_uid.append(each[1][0])
    uid2value[each[1][0]] = each[1][2]

path_submit = '/home/liyouru/DataSet/JDD/Loan_Forecasting.csv'
df2 = pd.read_csv(path_submit,header = 0,sep = ',')

ensemble_value = []
for each in df2.iterrows():
    if int(each[1][0]) in value_uid:
        temp = 0.4 * uid2value[int(each[1][0])] + 0.6 * float(each[1][1])
        ensemble_value.append(max(0,float(temp)))
    else:
        ensemble_value.append(0)
print len(ensemble_value),ensemble_value[0]
```
